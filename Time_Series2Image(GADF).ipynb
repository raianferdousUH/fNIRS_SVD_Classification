```python
from mpl_toolkits.axes_grid1 import ImageGrid
from pyts.image import GramianAngularField
from pyts.datasets import load_gunpoint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# HbO Conversion from Signal to Image
subj = 13
stim = 3


#data = pd.read_csv('Sub'+str(subj)+'_Visual_'+str(stim)+'_HbO.csv',header= None)
data_hbo = pd.read_csv('Sub'+str(subj)+'_Motor_'+str(stim)+'_HbO.csv',header= None)
```

```python
col = len(data_hbo.columns)
col
```

```python
# Perform Gramian Angular Field Difference to the siganl and convert them to images and save the images
for i in range(col):
    df= data_hbo.iloc[: , i]
    df= df.values.reshape(1,-1)
    gaf = GramianAngularField(method= 'difference')
    
    im = gaf.transform(df)
    new_image = plt.imshow(im[0])
    fig3 = plt.gcf()
    # show the plot
    #plt.show()
    #Save figure 
    #fig3.savefig('/Users/talukdarraianferdous/Documents/fNIRS Data/CPGM/GAFD_CNN/Images/Sub'+str(subj)+'/Visual/Sub'+str(subj)+'_Visual'+str(stim)+'_'+str(i)+'_HbO', dpi=500)
    #fig3.savefig('/Users/talukdarraianferdous/Documents/fNIRS Data/CPGM/GAFD_CNN/Images/Sub'+str(subj)+'/Motor/Sub'+str(subj)+'_Motor'+str(stim)+'_'+str(i)+'_HbO', dpi=500)
    
    plt.show()
    i += 1
    
```
