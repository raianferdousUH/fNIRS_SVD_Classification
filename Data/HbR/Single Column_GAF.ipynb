```python
from mpl_toolkits.axes_grid1 import ImageGrid
from pyts.image import GramianAngularField
from pyts.datasets import load_gunpoint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
subj = 13
stim = 3


data = pd.read_csv('Sub'+str(subj)+'_Visual_'+str(stim)+'_HbR.csv',header= None)
#data = pd.read_csv('Sub'+str(subj)+'_Motor_'+str(stim)+'_HbR.csv',header= None)
```

```python
col = len(data.columns)
col
```

```python
for i in range(col):
    df= data.iloc[: , i]
    df= df.values.reshape(1,-1)
    gaf = GramianAngularField(method= 'difference')
    
    im = gaf.transform(df)
    new_image = plt.imshow(im[0])
    fig3 = plt.gcf()
    # show the plot
    #plt.show()
    #Save figure 
    fig3.savefig('/Users/talukdarraianferdous/Documents/fNIRS Data/CPGM/GAFS_CNN/HbR/Visual/Sub'+str(subj)+'_Visual'+str(stim)+'_'+str(i)+'_HbR', dpi=500)
    #fig3.savefig('/Users/talukdarraianferdous/Documents/fNIRS Data/CPGM/GAFS_CNN/HbR/Motor/Sub'+str(subj)+'_Motor'+str(stim)+'_'+str(i)+'_HbR', dpi=500)
    
    plt.show()
    i += 1
```
